```python
import  matplotlib.pyplot as plt ###畫圖###
import pandas as pd  
import numpy as np
a = list(range(1,101))
df = pd.DataFrame(np.random.randn(100, 4), index = a , columns = list("ABCD"))
```

```python
print(df)
plt.plot(df)
plt.plot(df, "o--b")
plt.show()
```

```python
plt.plot(a,df["B"],df["B"],label="B") #兩者散佈圖
plt.plot(a,df["C"],label="C")
plt.grid() #格線
plt.ylabel("foolish rate")
plt.legend(loc=1) #上面plot的地方加label，這裡用legend可顯示標籤
plt.show()

plt.plot(a, df["B"], a, df["C"], "g--o", label ="???") #一次show兩個
plt.legend(loc=1)
plt.show()
#print(plt.axis()) #顯示x軸範圍，也可自訂義想要的XY軸範圍 plt.axis([xmin, xmax, ymin, ymax]) (當然前面要自己先賦值四個參數)

plt.scatter(df["B"],df["C"]) #散佈圖
#plt.bar(,) 柱狀圖
```
