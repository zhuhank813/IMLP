```python
import pandas as pd

data = {
"名稱":["客戶A", "客戶B", "客戶A", "客戶D", "客戶C"],
"編號":["001", "002", "003", "004", "005"],
"數量":["3", "4", "4", "8", "56"]}

df = pd.DataFrame(data) #將資料轉成dataframe
df.groupby("名稱").sum()#對有數值得地方做分組加總
```

```python
import pandas as pd

data = {
"channel":["網路", "網路", "電視", "電視", "郵購"],
"company":["001", "002", "003", "004", "005"],
"num":[3, 4, 4, 8, 56]}

ordinals = ["A","B","C","D","E"]

df2 = pd.DataFrame(data, index = ordinals)#將index設成自己想要的樣子

score = ["good","bad","good","good","bad"]
df2["score"] = score #增加欄位

df2
```

```python
pivot_tb = df2.pivot_table(index = "channel", columns="company", values="num") #樞紐分析
pivot_tb
```
