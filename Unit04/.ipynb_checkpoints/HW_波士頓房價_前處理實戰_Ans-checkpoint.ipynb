# 步驟流程

### 使用實戰2切割的數據完成以下

from sklearn import linear_model<br>
linear_model.LinearRegression()<br>

<br>* 輸出模型參數、評價模型<br>
此題主要使用線性迴歸<br>
要解決的問題是如何量化地找出這些特徵值和目標值（即房價）的線性關係<br>
MEDV = k1*CRIM + k2*ZN + … + k13*LITAT + b

<br>* 獲取測試集房價的預測結果，歸納出最終結論
xxx.predict(test_X)
![預測模型](lr_out.png)


```python
from sklearn import datasets
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


'''----------load 數據集-----------'''
dataset = datasets.load_boston()

# x 訓練特征：['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS','RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
x = dataset.data

target = dataset.target
#把label變為(?, 1)維度，為了使用下面的數據集合分割
y = np.reshape(target,(len(target), 1))

#將數據集1:3比例分割為 測試集：訓練集  設定random_state=1
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)


'''----------定義線性回歸模型，進行訓練、預測-----------'''
lr = linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)

print("y_test\n",y_test)

'''----------圖形化預測結果-----------'''
#只顯示前50個預測結果，太多的話看起來不直觀
plt.xlim([0,50])
plt.plot( range(len(y_test)), y_test, 'r', label='y_test')
plt.plot( range(len(y_pred)), y_pred, 'g--', label='y_predict' )
plt.title('sklearn: Linear Regression')
plt.legend()
plt.savefig('lr_out.png')
plt.show()


'''----------輸出模型參數、評價模型-----------'''
# 使用metrics.XXX()函式、()內參數放入預測值與目標值 
print("斜率：",lr.coef_)
print("截距：",lr.intercept_)
print("MAE:",metrics.mean_absolute_error(y_pred,y_test))
print("MSE:",metrics.mean_squared_error(y_test,y_pred))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))

#輸出模型對應R-Square
print(lr.score(x_train,y_train))
print(lr.score(x_test,y_test))


 
```
